```python
from apify_client import ApifyClient

# Initialize the ApifyClient with your API token
# SAYA MENYENSOR TOKEN ANDA DEMI KEAMANAN. Ganti dengan token asli Anda.
client = ApifyClient("apify_api_xxxxxxxxxxxxxxxxxxxxxxxxxxxxx")

# Prepare the Actor input
run_input = {
    "commentsPerPost": 2300,
    "excludePinnedPosts": False,
    "maxRepliesPerComment": 5,
    "postURLs": [
        "https://vt.tiktok.com/ZSP19V29E/"
    ],
    "resultsPerPage": 100
}

print("Sedang menjalankan Actor...")

# Run the Actor and wait for it to finish
run = client.actor("BDec00yAmCm1QbMEI").call(run_input=run_input)

# PERBAIKAN DI SINI: Gunakan ["defaultDatasetId"] bukan {"defaultDatasetId"}
dataset_id = run["defaultDatasetId"]

print(f"Actor selesai. Mengambil data dari Dataset ID: {dataset_id}")

# Mengambil item dari dataset
items = list(client.dataset(dataset_id).iterate_items())

# Contoh: Menampilkan jumlah data yang didapat
print(f"Berhasil mengambil {len(items)} komentar.")

# Opsional: Print item pertama untuk memastikan data masuk
if items:
    print(items[0])
```
